1. What is the problem you are solving?
> Businesses or your audience care about problems. Make the problem clear and relatable

Kenya's government, telecoms, and NGOs make decisions about internet infrastructure investment — but they often work from national averages that hide massive county-level inequality. A national figure of 35% sounds like reasonable progress yet it hide the fact that West Pokot sits at 9.1% while Nairobi is at 64.7% — a 55-point gap between counties in the same country.
The problem is: decisions made on national averages leave the most disconnected counties further behind. No one can fix a gap they can't see clearly.

Mobile phone usage nationally is 64.9%. Internet use is 35.0%. That 29.9-point gap means roughly 30% of Kenyans have a phone but are not using mobile data.


The gender gap is worse in remote counties.
Nationally, internet use is 37.8% for men vs 32.2% for women — a 5.6-point gap. But in counties like Samburu (25.4% male vs 12.5% female), Mandera (32.5% vs 17.5%), and Uasin Gishu (50.2% vs 34.1%), the gap widens dramatically. Women in remote counties are doubly excluded, by geography and by gender.
